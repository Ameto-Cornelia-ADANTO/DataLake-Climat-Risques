```python
from pyspark.sql import SparkSession
import requests
import os

spark = SparkSession.builder \
    .appName("NOAA_Ingestion") \
    .getOrCreate()

# ---------------------------
# 1. Téléchargement fichiers NOAA
# ---------------------------

# Exemple dataset NOAA (température, vent, pluie)
url = "https://www.ncei.noaa.gov/data/global-summary-of-the-day/access/2023/FRANCE.csv"

raw_folder = "/raw/noaa/2023/" 
os.makedirs("noaa_temp", exist_ok=True)
local_file = "noaa_temp/noaa_2023.csv"

print("Téléchargement du fichier NOAA...")
r = requests.get(url)
open(local_file, "wb").write(r.content)
print("Fichier téléchargé !")

# ---------------------------
# 2. Lecture avec Spark
# ---------------------------

df_noaa = spark.read.csv(local_file, header=True, inferSchema=True)
df_noaa.show(5)

# ---------------------------
# 3. Sauvegarde RAW
# ---------------------------

df_noaa.write.mode("overwrite").parquet("hdfs://namenode:8020/raw/noaa/2023/data.parquet")

print("Données NOAA déposées en RAW.")
```
